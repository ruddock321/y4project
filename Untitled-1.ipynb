```python
import h5py
import matplotlib.pyplot as plt
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F
from torch.utils.data import DataLoader
import torchvision.datasets as datasets
import torchvision.transforms as transforms

file_path = 'C:\\Users\\Dell\\Downloads\\Garstec_AS09_chiara.hdf5'
```

```python
with h5py.File(file_path, 'r') as hdf:

    '''
    We want to make a HR diagram plot with a sample of mass (1.0 < M < 1.1) 
    solar masses, and alpha_mlt initially set to ~1.79.
    '''

    # Navigate through Grid -> tracks -> track00001
    grid = hdf['grid']
    tracks = grid['tracks']

    # Initialize a list to hold the tracks that meet our desired criteria
    selected_tracks = []

    for track_name in tracks:

        track = tracks[track_name]

        # Check if 'massini' dataset exists and is not empty
        if 'massini' in track and track['massini'].size > 0:
            massini_value = track['massini'][()][0]  # Get the first element if it exists
        else:
            massini_value = 0  # Put outside of our desired range so it isn't selected

        # Check if 'alphaMLT' dataset exists and is not empty
        if 'alphaMLT' in track and track['alphaMLT'].size > 0:
            alphaMLT_value = track['alphaMLT'][()][0]  # Get the first element if it exists
        else:
            alphaMLT_value = 0  # Put outside of our desired range so it isn't selected

        # Appending the track list to contain desired tracks
        
        if all([
            1.0 < massini_value < 1.1,
            1.785 < alphaMLT_value < 1.795
        ]):  
            selected_tracks.append(track)


    for track in selected_tracks:
        # Access effective temperature and luminosity as np arrays
        teff_data = track['Teff'][:] 
        luminosity_data = track['LPhot'][:]
        age_data = track['age'][:]

        track_name = track.name.rsplit('/', 1)[-1]  # Split from the right, keep the last part

        # convert Teff, LPhot and age to Pytorch sensors

        Teff_tensor = torch.FloatTensor(teff_data).view(-1, 1)
        LPhot_tensor = torch.FloatTensor(luminosity_data).view(-1, 1)
        age_tensor = torch.FloatTensor(age_data).view(-1, 1)
            
    print(len(selected_tracks))      # checking how many tracks are within requirements
```

```python
# Define neural network
class StellarTracksModel(nn.Module):
    def __init__(self):
        super(StellarTracksModel, self).__init__()
        self.input_layer = nn.Linear(1, 128)  # Input layer to hidden layer
        self.hidden_layer2 = nn.Linear(128, 64)
        self.output_layer = nn.Linear(64, 2)  # Hidden layer to output layer
        self.activation = nn.Tanh()            # Activation function

    def forward(self, x):
        x = self.activation(self.input_layer(x))  # Apply hidden layer and activation
        x = self.activation(self.hidden_layer2(x))
        x = self.output_layer(x)                    # Apply output layer
        return x
    

# Train the model
# Initialize the model, loss function, and optimizer
model = StellarTracksModel()
criterion = nn.MSELoss()  # Mean Squared Error loss
optimizer = optim.Adam(model.parameters(), lr=0.1)  # Adam optimizer



# Training loop
num_epochs = 1000
for epoch in range(num_epochs):
    model.train()  # Set the model to training mode
    optimizer.zero_grad() 

    # Forward pass
    outputs = model(age_tensor)
    Teff_pred = outputs[:, 0]
    LPhot_pred = outputs[:, 1]


    loss1 = criterion(Teff_pred, Teff_tensor)  
    loss2 = criterion(LPhot_pred, LPhot_tensor)
    total_loss = loss1 + loss2

    # Backward pass and optimization
    total_loss.backward()
    optimizer.step()
    

    # Print the loss every 100 epochs
    if (epoch + 1) % 100 == 0:
        print(f'Epoch [{epoch + 1}/{num_epochs}], Loss: {total_loss.item():.4f}')
```

```python
model.eval()
with torch.no_grad():
    predictions = model(age_tensor)
    LPhot_pred = predictions[:, 0].numpy()  # Predicted LPhot values
    Teff_pred = predictions[:, 1].numpy()   # Predicted Teff values

age_np = age_tensor.numpy()
plt.figure(figsize=(10, 5))

print(Teff_pred, LPhot_pred)

plt.plot(Teff_pred, LPhot_pred, label='Predicted LPhot', color='cyan')
plt.plot(Teff_tensor.numpy(), LPhot_tensor.numpy(), label='True LPhot', color='blue')

plt.title('Predictions of LPhot and Teff')
plt.gca().invert_xaxis() #Invert x-axis to have proper form of HR diagram
plt.xlabel('Age (not shown in plot)')
plt.ylabel('Output Values')
plt.legend()
plt.grid()
plt.show()
```
